### Functions for processing daily data

Merge column names with daily dataset

```python
def export_daily_event(filename):
    colsnames = pd.read_csv("header_fieldids_for_daily_data.csv")
    event = pd.read_csv(filename, sep="\t", names=colsnames["Field Name"].values.tolist())
    filename_export = filename.split(".")[0]+"event.csv"
    event.to_csv(filename_export, index=False)
    return
```

```python
export_daily_event("20160223.export.csv")
```

Merge several daily datasets

```python
# Input is a list of file names or YYYYMMDDtoYYYYMMDD, e.g ["20160215event.csv","20160216event.csv"] or "20160215to20160221"
# Please put the oldest date at the first and the nearest date at the last
def merge_several_daily(listoffilename):
    if isinstance("listoffilename",list) or isinstance("listoffilename",str):
        if isinstance("listoffilename",str):
            rangeoffilename = listoffilename.split("to")
            listoffilename = []
            for i in range(int(rangeoffilename[0]),int(rangeoffilename[1])+1):
                listoffilename.append(str(i)+"event.csv")
        listoffile = []
        for filename in listoffilename:
            listoffile.append(pd.read_csv(filename)) 
        merged = pd.concat(listoffile,ignore_index=True)
        print str(listoffilename)+"is concated"
        #merged.append([pd.read_csv(filename) for filename in listoffile],ignore_index=True)
        filename_export = listoffilename[0][:8]+"to"+listoffilename[-1][:8]+"event.csv"
        merged.to_csv(filename_export, index=False)
    else:
        print "Please check input"
    return
```

```python
merge_several_daily(["20160215event.csv","20160216event.csv"])
```

```python
merge_several_daily("20160215to20160221")
```

### Process other support dataset

Split event_codebook into root level and base level

```python
eventcodebook = pd.read_csv("Event_codebook.csv")
```

```python
eventrootcodebook = eventcodebook[eventcodebook.apply(lambda x: x['EVENTDESCRIPTION'] == x['EVENTDESCRIPTION'].upper(), axis=1)]
eventrootcodebook[["EVENTDESCRIPTION"]] = eventrootcodebook["EVENTDESCRIPTION"].apply(lambda x:x[0]+x[1:].lower())
```

```python
eventrootcodebook
```

```python
eventrootcodebook.to_csv("EventRoot_codebook.csv", index=False)
eventbasecodebook = eventcodebook[eventcodebook.apply(lambda x: x['EVENTDESCRIPTION'] <> x['EVENTDESCRIPTION'].upper(), axis=1)]
eventbasecodebook.to_csv("EventBase_codebook.csv", index=False)
```
